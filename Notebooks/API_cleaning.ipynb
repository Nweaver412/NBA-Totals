```python
from nba_api.stats.endpoints import leaguegamefinder
from datetime import datetime, timedelta
import pandas as pd

# Define the date range
start_date = datetime(2023, 1, 1)  # Replace with your desired start date
end_date = datetime(2023, 12, 31)  # Replace with your desired end date

# Initialize an empty DataFrame to store the results
all_games_data = pd.DataFrame()

# Loop through the date range
current_date = start_date
while current_date <= end_date:
    formatted_date = current_date.strftime('%Y-%m-%d')
    
    try:
        # Make a request for home games
        home_games_response = leaguegamefinder.LeagueGameFinder(date_from_nullable=formatted_date, date_to_nullable=formatted_date, location_nullable='Home')
        home_games_data = home_games_response.get_data_frames()[0]

        # Make a request for away games
        away_games_response = leaguegamefinder.LeagueGameFinder(date_from_nullable=formatted_date, date_to_nullable=formatted_date, location_nullable='Away')
        away_games_data = away_games_response.get_data_frames()[0]

        # Combine home and away games data
        all_games_data = all_games_data.append(home_games_data, ignore_index=True)
        all_games_data = all_games_data.append(away_games_data, ignore_index=True)

    except Exception as e:
        print(f"Error for date {formatted_date}: {str(e)}")

    # Move to the next date
    current_date += timedelta(days=1)

# Display the combined data
print(all_games_data)
```
