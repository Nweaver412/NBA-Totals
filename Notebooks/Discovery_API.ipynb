```python
import pandas as pd
from datetime import datetime, timedelta
from nba_api.stats.static import teams
from nba_api.stats.endpoints import teamgamelog

# Find team by abbreviation
team_abbreviation = 'LAL'  # Example: Los Angeles Lakers
nba_teams = teams.get_teams()
team_dict = [team for team in nba_teams if team['abbreviation'] == team_abbreviation][0]

# Fetch team game log data
team_id = team_dict['id']
current_season = '2023-24'  # Adjust the season accordingly
game_log = teamgamelog.TeamGameLog(team_id=team_id, season=current_season)

# Convert to DataFrame
game_log_df = game_log.get_data_frames()[0]

# Convert game date to datetime and filter for the last 10 days
game_log_df['GAME_DATE'] = pd.to_datetime(game_log_df['GAME_DATE'])
last_10_days = datetime.now() - timedelta(days=10)
filtered_games_df = game_log_df[game_log_df['GAME_DATE'] > last_10_days]

# Display the filtered DataFrame
filtered_games_df
```
